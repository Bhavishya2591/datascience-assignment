```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv("tesla_deliveries_dataset_2015_2025.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates removed successfully")
```

```python
# Graph 1: Tesla Deliveries Over Years

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Estimated_Deliveries')
plt.title("Tesla Deliveries Over Years")
plt.show()
```

```python
# production Trend

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Year', y='Production_Units')
plt.title("Tesla Production Trend")
plt.show()
```

```python
# Deliveries by region 

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Region', y='Estimated_Deliveries')
plt.title("Tesla Deliveries by Region")
plt.xticks(rotation=45)
plt.show()
```

```python
# deliveries by model

plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Model', y='Estimated_Deliveries')
plt.title("Tesla Deliveries by Model")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

## Insights

1. Tesla deliveries have increased steadily over time.
2. Production units show significant growth.
3. Model 3 and Model Y contribute strongly to deliveries.
4. North America and Europe show strong market performance.
5. Production and deliveries are positively correlated.


# Conclusion

The analysis shows Tesla experienced strong growth in deliveries and production between 2015 and 2025.

Key findings:
- Deliveries increased over the years
- Production capacity expanded significantly
- Model 3 and Model Y performed strongly
- Regional demand remained high in North America and Europe

This indicates Tesla’s growing global EV market dominance.
